## Baseline Model Results

The supplied CNN model was executed on the chest X-ray dataset containing three classes: BACTERIAL, NORMAL, and VIRAL. The dataset loader reported 4336 training images, 1083 validation images, and 437 test images.

The baseline model was trained for 8 epochs. The final training accuracy reached 0.8808, while the final validation accuracy was 0.7636. Test accuracy was 0.7597.

The accuracy curves suggest signs of overfitting, because training accuracy continued to improve while validation accuracy stopped improving and declined slightly in later epochs.

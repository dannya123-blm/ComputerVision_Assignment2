The model achieved a test accuracy of approximately 0.75 (75.05%).

```python
import os
import pandas as pd

train_path = r'F:\ComputerVision_Assignment2\chest_xray\train'
test_path = r'F:\ComputerVision_Assignment2\chest_xray\test'

def count_images(folder):
    counts = {}
    for class_name in os.listdir(folder):
        class_folder = os.path.join(folder, class_name)
        if os.path.isdir(class_folder):
            counts[class_name] = len([
                f for f in os.listdir(class_folder)
                if os.path.isfile(os.path.join(class_folder, f))
            ])
    return counts

train_counts = count_images(train_path)
test_counts = count_images(test_path)

print("Train:", train_counts)
print("Test:", test_counts)
```

## Overfitting Analysis

The model shows signs of overfitting.

Training accuracy increases steadily from approximately 0.72 to 0.88 over the 8 epochs. However, validation accuracy peaks at around 0.78 and then begins to decrease slightly in later epochs.

This indicates that the model is learning the training data well but is not generalising effectively to unseen validation data.

The gap between training and validation accuracy suggests that the model is overfitting.

To address overfitting, techniques such as using GlobalAveragePooling2D instead of Flatten, adding regularisation, or applying data augmentation can be used.

## Data Augmentation

To reduce overfitting, data augmentation techniques were applied.

Specifically, random horizontal flipping and small rotations were introduced to the training images.

This increases the diversity of the dataset and helps the model generalise better.

After applying data augmentation, the model was retrained.

The test accuracy decreased slightly from 0.7597 to 0.7345.

However, the validation accuracy became more stable, indicating reduced overfitting.

This suggests that the model is learning more general features rather than memorising the training data.
